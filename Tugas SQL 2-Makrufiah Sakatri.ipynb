##nomer 4, Tampilkan informasi customer yang mempunyai payment amount diatas 8000 dan urutkan berdasarkan nama customr sesuai abjad

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Membuat Query SQL
result  = connection.cursor()
command = """
             SELECT * FROM customer WHERE PAYMENT_AMT > '8000.00' ORDER BY CUST_NAME ASC;
          """

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

##nomer 5, Tampilkan data agensi yang tidak USA di london dan memiliki grade lebih dari 1

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Membuat Query SQL
result  = connection.cursor()
command = """
             SELECT * FROM customer WHERE CUST_COUNTRY NOT IN ('USA') AND GRADE > '1';
          """

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```
